```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('CarPrice_Assignment.csv')
df = pd.DataFrame(data)
df[['company', 'model']] = df['CarName'].str.split(' ', n =1, expand=True)
df
```

```python
X_train = df.drop(['car_ID','CarName','model','price'], axis = 1)
X_train
```

```python
y = df['price']
y_train = pd.DataFrame(y)
y_train
```

```python
X_train = pd.get_dummies(X_train, columns = ['symboling','company', 'fueltype','aspiration', 'doornumber','carbody','drivewheel','enginelocation','enginetype','cylindernumber','fuelsystem'])
X_train.head()
```

```python
def Z_normalise(X_train):
    mu = X_train.mean(axis = 0)
    std = np.std(X_train)
    X_train = ((X_train - mu)/std)
    return X_train
```

```python
X_train1= pd.DataFrame(Z_normalise(X_train))
X_train1
m = len(X_train1)
X_train1
```

```python
def normal_equation(X, y):
  """
  Calculates the coefficients of a linear regression model using the normal equation.

  Parameters:
    X (numpy.ndarray): The design matrix (features).
    y (numpy.ndarray): The target variable.

  Returns:
    numpy.ndarray: The coefficients of the linear regression model.
  """

  # Add a column of ones for the intercept term
  X_b = np.c_[np.ones((X.shape[0], 1)), X]

  # Calculate the coefficients using the normal equation
  theta = np.linalg.pinv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

  return theta
```

```python
X_train = X_train1.to_numpy()
y_train = y_train.to_numpy()
theta= normal_equation(X_train, y_train)
m,_ = X_train.shape
print(theta)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Predict using matrix multiplication for all samples at once
pred = np.dot(X_train, theta[1:])+theta[0]

# Compute root mean squared error (RMSE)
errors = mean_squared_error(y_train, pred, squared=False)

# Normalize by the mean of y_train
nrmse = errors / np.mean(y_train)

# Print the result with formatted output
print(f'NRMSE: {nrmse:.3f}')
```

```python
from sklearn.metrics import r2_score  
r2 = r2_score(y_train,pred) 
print('r2 score for model is', 100* r2) 
```
