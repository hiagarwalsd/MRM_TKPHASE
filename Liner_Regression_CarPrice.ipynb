```python
X_train = df.drop(['car_ID','CarName','model','price'], axis = 1)
X_train
```

```python
y = df['price']
y_train = pd.DataFrame(y)
y_train
```

```python
X_train = pd.get_dummies(X_train, columns = ['symboling','company', 'fueltype','aspiration', 'doornumber','carbody','drivewheel','enginelocation','enginetype','cylindernumber','fuelsystem'])
X_train.head()
```

```python
def Z_normalise(X_train):
    mu = X_train.mean(axis = 0)
    std = np.std(X_train)
    X_train = ((X_train - mu)/std)
    return X_train
```

```python
X_train1= pd.DataFrame(Z_normalise(X_train))
X_train1
m = len(X_train1)
X_train1
```

```python
def compute_cost(X, y, w, b):
    m = X.shape[0]
    cost = 0.0
    for i in range(m):
        f_wb_i = np.dot(w, X[i])+b
        cost +=(f_wb_i - y[i]) ** 2
    cost /= (2 * m)
    return cost
```

```python
def compute_gradient(X, y, w, b):
    m, n = X.shape
    dj_dw = np.zeros((n,))
    dj_db = 0
    for i in range(m):
        err = np.dot(X[i], w) + b - y[i]
        for j in range(n):
            dj_dw[j] += err*X[i,j]
        dj_db += err
    dj_dw /= m
    dj_db /= m
    return dj_db, dj_dw
```

```python
def gradient_descent(X, y, w_in, b_in, alpha, num_iter):
    J_history = []
    w = w_in
    b = b_in
    for i in range(num_iter):
        dj_db, dj_dw = compute_gradient(X, y, w, b)
        w = w - alpha * dj_dw
        b = b - alpha * dj_db
        if i<1000000:
            J_history.append(compute_cost(X, y, w, b))
    return w, b, J_history
```

```python
X_train = X_train1.to_numpy()
y_train = y_train.to_numpy()
initial_w = np.zeros(85)
initial_b = 0.0
iterations = 2500
alpha = 9.0e-3
w_final, b_final, J_hist = gradient_descent(X_train, y_train, initial_w, initial_b,alpha, iterations)
m,_ = X_train.shape
print(w_final, b_final)
```

```python
import matplotlib.pyplot as plt
fig, (ax1) = plt.subplots(1, constrained_layout=True, figsize=(12, 4))
ax1.plot(J_hist)
ax1.set_title("Cost vs. iteration")
ax1.set_ylabel('Cost')              
ax1.set_xlabel('iteration step')   
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Predict using matrix multiplication for all samples at once
pred = np.dot(X_train, w_final) + b_final

# Compute root mean squared error (RMSE)
errors = mean_squared_error(y_train, pred, squared=False)

# Normalize by the mean of y_train
nrmse = errors / np.mean(y_train)

# Print the result with formatted output
print(f'NRMSE: {nrmse:.3f}')
```

```python
from sklearn.metrics import r2_score  
r2 = r2_score(y_train,pred) 
print('r2 score for model is', 100* r2) 
```
